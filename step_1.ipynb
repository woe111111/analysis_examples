## 读取非标准但是工作中常用的json格式文本文件

```python
import json
data = []
with open("./data/json.txt","r") as f :
    for i in f.readlines():
        item = json.loads(i)
        data.append(item)
data
```

```python
df = pd.DataFrame(data)
df
```
